```python
ch1=str(input("donner un chaine: "))
ch2=str(input("donner un chaine: "))
a=ch1[:2]
b=ch2[:2]
print(a)
print(b)
result=b+ch1[2:]+a+ch2[2:]
print(result)
```

```python
ch1=str(input("donner un chaine: "))
n=int(input("donner l'index n: "))
result=(ch1[:n]+ch1[n+1:])
print(result)
```

```python
ch1=str(input("donner un chaine: "))
a=ch1[:1]
b=ch1[-1:]
print(a)
print(b)
result=b+ch1[1:-1]+a
print(result)
```
