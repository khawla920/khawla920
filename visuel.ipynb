```python
import matplotlib.pyplot as plt
activities = ['manger', 'dormir', 'travailler', 'divertissement']
slices = [5, 2, 6, 4]
colors = ['r', 'y', 'g', 'b']
plt.pie(slices, labels = activities, colors=colors,
startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
radius = 1.2, autopct = '%1.1f%%')
plt.legend()
plt.show()
```
